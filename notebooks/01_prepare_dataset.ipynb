```
from pathlib import Path
from protein_patch.data.fetch import sample_random_pdb_ids, fetch_pdbs
from protein_patch.data.prep import prep_dataset
from protein_patch.spec import PatchSpec

CACHE = Path('/content/pdb_cache')
OUT = Path('/content/drive/MyDrive/protein_patches')
ids = sample_random_pdb_ids(50, seed=42)
fetch_pdbs(ids, CACHE, delay=0.5)
summary = prep_dataset(CACHE, OUT, PatchSpec(), seed=42, val_fraction=0.2)
print(summary['n_patches'], 'patches ->', OUT)
```
